```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def load_scores(filename, average_window=100):
    with open(filename, 'r') as f:
        scores = list(map(int, f.readlines()))
        avg_score = []
        for i in range(len(scores) - average_window):
            avg_score += [np.mean(scores[i:i+average_window])]
    
        return scores, avg_score
```

```python
def get_success_index(list_of_scores):
    for idx, score in enumerate(list_of_scores):  # naive implementation
        if score >= 13.0:
            return idx
```
