```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
FILENAME = 'resnet_scores.txt'

with open(FILENAME, 'r') as f:
    scores = list(map(int, f.readlines()))
    
avg_score = []
for i in range(len(scores) - 100):
    avg_score += [np.mean(scores[i:i+100])]
```

```python
plt.plot(avg_score)
```
